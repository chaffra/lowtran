```python
import lowtran
from lowtran.plot import horiz
import pylab as pb
m = 1.0
km = 1000*m
deg = 1.0
nm = 1e-9*m
μm = 1e-6*m
cm = 1e-2*m
```

```python
observer_altitude = 10*km
target_range = 100*km
zenith_angle = 40.0*deg
wlshort = 100*nm
wllong = 20*μm
wlstep = 20.0/cm

c1 = {
        "zmdl": observer_altitude/km,
        "h1": observer_altitude/km,
        "range_km": target_range/km,
        "wlshort": wlshort/nm,
        "wllong": wllong/nm,
        "wlstep": wlstep/(1.0/cm),
    }

T_atm = lowtran.horiztrans(c1).squeeze()
horiz(T_atm,c1)
#pb.show()
```
